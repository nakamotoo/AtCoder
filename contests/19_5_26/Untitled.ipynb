```python
bits = [format(i,'b').zfill(2) for i in range(2**2)]

count = 0
for bit in bits:
    d = 1
    for i in range(2):
        sum = 0
        for sl in s[i]:
            sum = sum + int(bit[sl-1])
       
        if sum % 2 != p[i]:
            d = 0
            break
    count += d

print(count)
```
